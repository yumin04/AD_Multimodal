```python
import os
import pandas as pd
```

```python
# Path to your directory containing pickle files
directory_path = "/home/mason/TFN_AD/patients_csv"

# Label mapping dictionary
label_mapping = {'CN': 0, 'MCI': 1, 'Dementia': 2}
```

```python
# Iterate through all files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".pkl"):  # Process only pickle files
        file_path = os.path.join(directory_path, filename)
        
        # Load the pickle into a pandas DataFrame
        patient = pd.read_pickle(file_path)
        
        # Add the new column based on the mapping
        patient['integer_label'] = patient['DX'].map(label_mapping)
        
        # (Optional) Save the updated DataFrame back to the pickle file
        patient.to_pickle(file_path)
        
        print(f"Processed and updated: {filename}")
```
