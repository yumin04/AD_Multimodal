## Get All MRIs

```python
adni_images = "ADNI_IMG_32.5%_x"
image_folder = 'ADNI_IMG_32.5%_x'
csv_file = 'data/ADNI1_Screening_1.5T_6_23_2023.csv'

# csv_file = os.path.join(os.getcwd(),csv_file)

data = pd.read_csv(csv_file)

all_pngs = misc.list_files(image_folder, 'png')['all_file_paths']
missing_or_damaged_png_ids = []
images = []
for index, row in data.iterrows():
    image_path = []
    for filepath in all_pngs:
        if (row['Image Data ID'] + '') in filepath:
            image_path.append(filepath)
            break
    if not image_path:  # Check if image_path is empty
        continue  # Skip the current iteration
    image_path = image_path[0]
    try:
        Image.open(image_path)
    except:
        missing_or_damaged_png_ids.append(image_path.split('__')[1])
        all_pngs.remove(image_path)
        continue
    image = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
    image = tf.keras.preprocessing.image.img_to_array(image)
    images.append(image)

for i in missing_or_damaged_png_ids:
    data.drop(data[data['Image Data ID'] == i].index, inplace=True)
    
images = np.array(images)
images = images / 255.0

data['Image'] = list(images)
labels = data["Group"]
```

## Resnet50 Model

```python
BATCH_SIZE = 315
IMG_SIZE = (224, 224)
directory = "ADNI_IMG_32.5%_x_organized"
train_dataset = image_dataset_from_directory(directory,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             validation_split=0.2,
                                             subset='training',
                                             seed=20,
                                             label_mode = 'categorical')
validation_dataset = image_dataset_from_directory(directory,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             validation_split=0.2,
                                             subset='validation',
                                             seed=20,
                                             label_mode = 'categorical')
```

```python
base_model = tf.keras.applications.ResNet50(weights='imagenet', include_top = False)
# print(base_model.summary())
```

```python
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)

x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dense(512, activation='relu')(x)
preds = tf.keras.layers.Dense(3, activation ='softmax')(x)
```

```python
model = tf.keras.models.Model(inputs=base_model.input, outputs=preds)
```

```python
for layer in model.layers[:175]:
  layer.trainable = False
```

```python
for layer in model.layers[175:]:
  layer.trainable = True
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='accuracy', patience=5, restore_best_weights=True)
history = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=30,
    callbacks=[early_stopping]
)
```

```python
# Get training and validation accuracy and loss from the history object
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Plot training and validation accuracy
plt.figure()
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
```

```python
# Plot training and validation loss
plt.figure()
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
```

```python
# Get the final accuracy and loss for the last epoch
final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
final_train_loss = history.history['loss'][-1]
final_val_loss = history.history['val_loss'][-1]

# Print the final accuracy and loss
print("Final Training Accuracy:", final_train_acc)
print("Final Validation Accuracy:", final_val_acc)
print("Final Training Loss:", final_train_loss)
print("Final Validation Loss:", final_val_loss)
```

## Data Fusion (early fusion)

```python
data2 = pd.read_csv('data/ADNIMERGE.csv', low_memory=False)
```

```python
data2 = data2[['PTID', 'PTGENDER', 'AGE', 'PTRACCAT', 'DX_bl', "ADAS11","ADAS13","ADASQ4"]]
data2.sort_values('PTID')
data2.dropna(inplace=True)
```

```python
merged_data = pd.merge(data, data2, left_on='Subject', right_on='PTID', how='inner')

if merged_data.isnull().values.any():
    merged_data.dropna(inplace=True)
```

```python
seed_value = 42
tf.random.set_seed(seed_value)
```

```python
base_model = tf.keras.applications.ResNet50(weights='imagenet', include_top=False)
```

```python
image_input = Input(shape=(224, 224, 3))

n_numeric_features = 6
numeric_input = Input(shape=(n_numeric_features,))
```

```python
image_model = base_model(image_input)
image_model = tf.keras.layers.GlobalAveragePooling2D()(image_model)
```

```python
numeric_model = tf.keras.layers.Dense(128, activation='relu')(numeric_input)
```

```python
fusion_layer = Concatenate()([image_model, numeric_model])
fusion_layer = Dense(128, activation='relu')(fusion_layer)

n_classes = 3
output = Dense(n_classes, activation='softmax', name='output')(fusion_layer)
```

```python
combined_model = Model(inputs=[image_input, numeric_input], outputs=output)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)
combined_model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
merged_data = merged_data[["Sex","Age","Image","Group","ADAS11","ADAS13","ADASQ4"]]
```

```python
# Preprocess the numerical data (normalize Age, and convert Sex to dummy variable)
merged_data[['Age','ADAS11','ADAS13',"ADASQ4"]] = StandardScaler().fit_transform(merged_data[['Age','ADAS11','ADAS13',"ADASQ4"]])
merged_data = pd.get_dummies(merged_data, columns=['Sex','Group'])
```

```python
# Separate the features and labels
features = merged_data.drop(columns=['Group_AD', 'Group_MCI', 'Group_CN'])
labels = merged_data[['Group_AD', 'Group_MCI', 'Group_CN']]  # Assuming 3 classes for Group (AD, MCI, CN)
image_data = merged_data[['Image']]
```

```python
X_images_train, X_images_test, X_numeric_train, X_numeric_test, y_train, y_test = train_test_split(
    merged_data["Image"], merged_data[["Age","Sex_F","Sex_M","ADAS11","ADAS13","ADASQ4"]], labels, test_size=0.2, random_state=42
)
```

```python
import math

batch_size = 100  # Adjust the batch size based on your memory constraints

# Convert NumPy arrays to TensorFlow tensors in batches
num_samples_train = len(X_images_train)
num_batches_train = int(math.ceil(num_samples_train / batch_size))

num_samples_test = len(X_images_test)
num_batches_test = int(math.ceil(num_samples_test / batch_size))

X_images_train_tensors = []
X_images_test_tensors = []
X_numeric_train_tensors = []
X_numeric_test_tensors = []

# Convert training data
for i in range(num_batches_train):
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, num_samples_train)
    batch_images_train = np.array([image.flatten() for image in X_images_train[start_idx:end_idx]])
    X_images_train_tensors.append(tf.convert_to_tensor(batch_images_train.reshape(-1, 224, 224, 3), dtype=tf.float32))
    X_numeric_train_tensors.append(tf.convert_to_tensor(X_numeric_train[start_idx:end_idx], dtype=tf.float32))

# Convert test data
for i in range(num_batches_test):
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, num_samples_test)
    batch_images_test = np.array([image.flatten() for image in X_images_test[start_idx:end_idx]])
    X_images_test_tensors.append(tf.convert_to_tensor(batch_images_test.reshape(-1, 224, 224, 3), dtype=tf.float32))
    X_numeric_test_tensors.append(tf.convert_to_tensor(X_numeric_test[start_idx:end_idx], dtype=tf.float32))

# Concatenate the tensors to get the final X_images_train, X_images_test, X_numeric_train, and X_numeric_test
X_images_train = tf.concat(X_images_train_tensors, axis=0)
X_images_test = tf.concat(X_images_test_tensors, axis=0)
X_numeric_train = tf.concat(X_numeric_train_tensors, axis=0)
X_numeric_test = tf.concat(X_numeric_test_tensors, axis=0)
```

```python
# best hyperparameters: patience = 18
early_stopping = EarlyStopping(monitor='val_accuracy', patience=15, restore_best_weights=True)
```

```python
# best hyperparameters: epoches=150
history = combined_model.fit([X_images_train, X_numeric_train], y_train, epochs=150, batch_size=100, validation_data=([X_images_test, X_numeric_test], y_test), callbacks=[early_stopping])
```

```python
test_loss, test_accuracy = combined_model.evaluate([X_images_test, X_numeric_test], y_test, verbose=0)

print("Train Set:")
print("Accuracy:", history.history['accuracy'][-1])
print("Loss:", history.history['loss'][-1])

print("\nTest Set:")
print("Accuracy:", test_accuracy)
print("Loss:", test_loss)
```

```python
# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Get the predicted probabilities for the test data
y_pred_probs = combined_model.predict([X_images_test, X_numeric_test])

# Compute micro-averaged ROC curve and AUC
fpr_micro, tpr_micro, _ = roc_curve(y_test.to_numpy().ravel(), y_pred_probs.ravel())
roc_auc_micro = auc(fpr_micro, tpr_micro)

# Plot the micro-averaged ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr_micro, tpr_micro, color='blue', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc_micro))
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
