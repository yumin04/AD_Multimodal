```python
import os
import re

# Path to the folder containing the images
folder_path = "/home/mason/ADNI_Dataset/ADNI_IMG_32.5%_x"

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    # Match the "I" segment using a regex
    match = re.search(r'(I\d+)', filename)
    if match:
        # Extract the matched segment
        new_name = match.group(1) + ".png"
        # Build full old and new file paths
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} -> {new_name}")
```
